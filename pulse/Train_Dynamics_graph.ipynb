```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd
```

```python
test_loss = pd.read_csv('../tensorboard/ex_1/run-loss_test-tag-loss.csv', delimiter=',')
test_loss = test_loss[['Step', 'Value']]
test_loss = pd.DataFrame(test_loss)
test_loss['log(Value)'] = np.log(test_loss[['Value']])
test_loss
```

```python
train_loss = pd.read_csv('../tensorboard/ex_1/run-loss_training-tag-loss.csv', delimiter=',')
train_loss = train_loss[['Step', 'Value']]
train_loss = pd.DataFrame(train_loss)
train_loss['log(Value)'] = np.log(train_loss[['Value']])
train_loss
```

```python
df_test = test_loss.copy()
df_test['Выборка'] = 'Тестовая'
df_train = train_loss.copy()
df_train['Выборка'] = 'Тренировочная'
df = pd.concat([df_train,df_test])
df
```

```python
fig = px.line(df,x='Step',y='Value',color='Выборка')
fig.update_yaxes(range=[0.1, 0.3])
fig.update_xaxes(range=[40000, df[['Step']].iloc[-1]])
fig.update_layout(title="График зависимости потерь на тренировочной и тестовой выборке от количества батчей",
                  xaxis_title="Количество батчей",
                  yaxis_title="Потери",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
#px.scatter(train_loss,x="Step",y="Value")
# px.scatter(test_loss, x="Step",y="Value")
```
