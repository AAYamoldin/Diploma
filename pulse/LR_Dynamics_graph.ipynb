```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd
```

```python
test_lr0001 = pd.read_csv('../tensorboard/ex_1/run-loss_test-tag-loss.csv', delimiter=',')
test_lr0001['lr'] = 0.0001
test_loss = test_lr0001.copy()

test_lr0002 = pd.read_csv('../tensorboard/ex_2_LR0.0002/run-loss_test-tag-loss.csv', delimiter=',')
test_lr0002['lr'] = 0.0002
test_loss = pd.concat([test_loss, test_lr0002])

test_lr0004 = pd.read_csv('../tensorboard/ex_2_LR0.0004/run-loss_test-tag-loss.csv', delimiter=',')
test_lr0004['lr'] = 0.0004
test_loss = pd.concat([test_loss, test_lr0004])

test_lr0008 = pd.read_csv('../tensorboard/ex_2_LR0.0008/run-loss_test-tag-loss.csv', delimiter=',')
test_lr0008['lr'] = 0.0008
test_loss = pd.concat([test_loss, test_lr0008])

#test_lr0016 = pd.read_csv('../tensorboard/ex_2_LR0.0016/run-loss_test-tag-loss.csv', delimiter=',')
#test_lr0016['lr'] = 0.0016
#test_loss = pd.concat([test_loss, test_lr0016])

test_lr0032 = pd.read_csv('../tensorboard/ex_2_LR0.0032/run-loss_test-tag-loss.csv', delimiter=',')
test_lr0032['lr'] = 0.0032
test_loss = pd.concat([test_loss, test_lr0032])

fig = px.line(test_loss,x='Step',y='Value',color='lr')
fig.update_yaxes(range=[0.1, 0.5])
fig.update_xaxes(range=[0, 350000])
fig.update_layout(title="График зависимости потерь на тестовой выборке при различном learning rate",
                  xaxis_title="Количество батчей",
                  yaxis_title="Потери",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False)  # Turn off reloader if inside Jupyter
```
